## Polynomial functions

We start with **polynomial functions**, a group of functions, where the variable is subject to powers.

Consider time $t$ [s] as independent variable, then a **polynomial function** is given as:
$$
f(t) = \sum\limits_{i=0}^{n} a_i t^i
$$
with $a_i$ the coefficients of the terms.

We define the time scale, using the `linspace` function from the `numpy` package,
here from $t=0$s to $t=6$s, with 51 steps:

```python
t = np.linspace(0,6,51)
```

Next, we define the **polynomial function**:

```python
def fpoly(t,a):
    """
    calculate polynomial function
    f(t) = a0 + a1 t + a2 t^2 + ...
    input:
    t    - array of time values
    a    - array of coefficients
    returns:
    fpoly(t) - array of function values
    """
    fpoly=0
    n = len(a)
    for i in range(n):
        fpoly += a[i]*t**i
    return fpoly
```

As example, we caluclate a polynomial of order 2, with coefficients:
$a_0=10$, $a_1=5$, $a_2=0.8$

```python
a = [10,5,0.8]
#a = [1]
#a = [0,1]
#a = [0,0,1]

function = '$f(t)='
for i in range(len(a)):
    function += str(a[i])+'t^{'+str(i)+'}'
    if (i < len(a)-1): function += '+'
function += '$'

plt.figure(figsize=(10,4))
plt.title(function)
plt.xlabel('t')
plt.ylim([-20,100])
plt.plot(t,fpoly(t,a),lw=4,label='f$(t)$')
plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(10,4))
plt.xlabel('t')
plt.ylabel('y(t)')
plt.plot(t,fpoly(t,a=[1]),linewidth=4,color='blue',label='constant: $y(t)=y_0$')
plt.plot(t,fpoly(t,a=[1,1]),linewidth=4,color='red',label='linear: $y(t)=y_0+y_1 t$')
plt.plot(t,fpoly(t,a=[1,1,1]),linewidth=4,color='green',label='quadratic: $y(t)=y_0+y_1 t+y_2 t^2$')
plt.legend()
```

## Using `interact` from the `ipywidgets` library:

```python
def fpoly2(t,a0=1,a1=1,a2=1):
    """
    calculate polynomial function
    f(t) = a0 + a1 t + a2 t^2 + ...
    # version for interact ...
    """
    # assemble function   
    fpoly2  = a0*np.ones_like(t)
    fpoly2 += a1*t
    fpoly2 += a2*t**2
    # plot function
    plt.figure(figsize=(10,4))
    plt.xlabel('t')
    plt.ylim([-100,100])
    plt.plot(t,fpoly2,lw=4,label='f')
    plt.grid()
    plt.legend()
    return
```

```python
widgets.interact(fpoly2,
                 t=widgets.fixed(t),
                 a0=widgets.IntSlider(min=-50, max=50, step=1, value=1),
                 a1=widgets.IntSlider(min=-10, max=10, step=1, value=1),
                 a2=widgets.FloatSlider(min=-3, max=3, step=0.2, value=0))
```
