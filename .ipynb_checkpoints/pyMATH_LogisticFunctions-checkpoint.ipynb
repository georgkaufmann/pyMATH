# `pyMATH`

pyMATH, a program package for playing with math in python.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
#from ipywidgets import interact, interactive, fixed, interact_manual
import scipy.integrate
```

----
## Logistic functions

We continue with **logistic functions**, a group of functions, which combine
exponential growth and a long-term limit as threshold.

Consider time $t$ [s] as independent variable, then a **logistic function** is given as:
$$
f(t) = \frac{K}{1+ \left( \frac{K-P_0}{P_0} \right) e^{-rt}}
$$
with 
- $P_0$ the initial value,
- $r$ the growth rate [1/s], and
- $K$ the final population.

We define the time scale, using the `linspace` function from the `numpy` package,
here from $t=0$s to $t=6$s, with 51 steps:

```python
t = np.linspace(0,6,51)
```

Next, we define the **logistic function**:

```python
def fLogistic(t,P0,r=1,K=1):
    """
    calculate logistic function
    f(t) = ...
    input:
    t    - array of time values
    P0   - initial coefficient
    r,K  - coefficients
    returns:
    fLogistic(t) - array of function values
    """
    f = K/(1+((K-P0)/P0)*np.exp(-r*t))
    return f
```

We use 
- two settings for parameter values $P0=0.01$ and $P0=0.2$
- a parameter value $r=1.1$
- a parameter value $K=1$

```python
P0 = [0.01,0.2]
r  = 1.1
K  = 1

plt.figure(figsize=(10,4))
plt.title('Logistic function')
plt.xlabel('t')
plt.ylim([-0.1,1.1])
plt.plot(t,fLogistic(t,P0[0]),lw=4,label='P0='+str(P0[0]))
plt.plot(t,fLogistic(t,P0[1]),lw=4,label='P0='+str(P0[1]))
plt.grid()
plt.legend()
```

## Using `interact` from the `ipywidgets` library:

```python
def fLogistic2(t,P0,r=1,K=1):
    """
    calculate logistic function
    f(t) = ...  
    version for interact ...
    """
    # assemble function   
    fLogistic2 = K/(1+((K-P0)/P0)*np.exp(-r*t))
    # plot function
    plt.figure(figsize=(10,4))
    plt.xlabel('t')
    plt.ylim([-1,3])
    plt.plot(t,fLogistic2,lw=4,label='f')
    plt.grid()
    plt.legend()
    return
```

```python
widgets.interact(fLogistic2,
                 t=widgets.fixed(t),
                 P0=widgets.FloatLogSlider(min=-3, max=-1, step=0.1, value=-1),
                 r=widgets.FloatSlider(min=0.1, max=3, step=0.2, value=1.1),
                 K=widgets.FloatSlider(min=1, max=3, step=0.2, value=0))
```

## Differential equation

The logistic function 
$y(t)=\frac{y}{1+ ( \frac{K-y_0}{y_0} ) e^{-rt}}$ is the solution of the **differential equation**:
$$\fbox{$
\frac{dy}{dt} = r y \left( 1 - \frac{y}{K} \right)
$}$$
with $t$ time [s], 
$y$ the population number [-], 
$r$ the growth rate [1/s], and
$K$ the final population.

The differential equation can be solved, subject to the **initial condition**
$$\fbox{$
y(t=0)=y_0
$}$$

```python
#define coefficients
r = 1.1
K = 1

# define the right-hand side
def rhs(t,P,r,K):
    dPdt = r*P*(1-P/K)
    return dPdt

# define the initial condition
y0 = [0.01]

# solve the ODE for the given range
dydt = scipy.integrate.solve_ivp(rhs,[t.min(),t.max()],P0,args=(r,K),t_eval=t)
print(dydt.t,dydt.y)
```

```python
plt.figure(figsize=(12,6))
plt.plot(t,fLogistic(t,P0=P0[0],r=r,K=K),linewidth=10,color='gray',label='$f(t)=e^{-t}$')
plt.plot(t,fLogistic(t,P0=P0[1],r=r,K=K),linewidth=10,color='gray')
plt.plot(dydt.t,dydt.y[0],linewidth=2,color='orange',label='ode solution, P0='+str(P0[0]))
plt.plot(dydt.t,dydt.y[1],linewidth=2,color='red',label='ode solution, P0='+str(P0[1]))
plt.grid()
plt.legend()
```

## Normalized logistic equation: Sigmoid function

We normalize the logistic differential equation, using
- $K=1$
- $r=1$
and then set $P=x$:
$$\fbox{$
\frac{dx}{dt} = x \left( 1 - x \right)
$}$$
with $t$ time [s], 
$x$ the normalized population number [-].

The differential equation can be solved, subject to the **initial condition**
$$\fbox{$
x(t=0)=x_0
$}$$

The analytical solution is known as **sigmoid function**:
$$
x(t) = \frac{1}{1+e^{-t}}
$$

----
## Recurrence relation: Logistic map

We re-use the right-hand side of the logistic differential equation in the following
**recurrence relation**:
$$
x_{n+1} = r x_{n} \left( 1 - x_{n} \right)
$$
This relation is also called **logistic map**.

We write a function for the logistic map:

```python
def logisticMap(x0,r,nmax):
    xnew = [x0]
    for i in range(1,nmax):
        xnew.append(r*xnew[i-1]*(1-xnew[i-1]))
    return xnew

def plot_logisticmap(r):
    nmax = 200
    t = [t for t in range(nmax)]
    x0 = 0.01
    plt.figure(figsize=(10,4))
    plt.xlim(0,nmax)
    plt.ylim([-0.1,1.1])
    plt.plot(t,logisticMap(x0,r,nmax), color='red',linestyle='-',linewidth=1,label='r='+str(round(r,1)))
```

```python
widgets.interact(plot_logisticmap, 
                r=widgets.FloatSlider(min=0.80,max=4.0,step=0.1,value=0.7,description='r'),
                )
```
