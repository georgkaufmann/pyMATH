```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
```

----
## Ordinary differential equations

### First order

We start with an **ordinary** differential equation of first order:
$$
y' = -y
$$
with 
- $y(t)$ the function of the independent variable $t$, and 
- $y'$ the short-hand notation of $y'=\frac{\partial y}{\partial t}$.

This is an first-order ODE, because
- it depends on **one** variable,
- it has **one** derivative,
- it is **one** equation.

This ODE can be solved, subject to an **initial condition**:
$$
y(t=0) = y_0
$$

A **solution** of this ODE is the **exponential function**:
$$
y(t) = e^{-t}
$$
We test this with the `solve_ivp` solver from the `scipy` package:

```python
# set variable range
t = np.linspace(0,6,51)

# define the right-hand side
def rhs(t,y):
    dydt = -np.exp(-t)
    return dydt

# define the initial condition
y0 = [1]

# solve the ODE for the given range
dydt = scipy.integrate.solve_ivp(rhs,[t.min(),t.max()],y0,t_eval=t,args=())
print(dydt.t,dydt.y)
```

```python
plt.figure(figsize=(12,6))
plt.plot(t,np.exp(-t),linewidth=10,color='gray',label='$f(t)=e^{-t}$')
plt.plot(dydt.t,dydt.y[0],linewidth=2,color='orange',label='ode solution')
plt.grid()
plt.legend()
```
