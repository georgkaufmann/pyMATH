## Fourier transform
As function $f(x)$ can be approximated by a polynomial:
$$
f(x) \simeq P(x)
$$
As a classical polynomial for expansion, we use the **Taylor series**, $P(x)=T_n(x,x_0)$, defined as:
$$
T_n(x,x_0) 
= \sum\limits_{n=0}^{\infty} \frac{f^{(n)}(x_0)}{n!} (x-x_0)^n
$$
Here, $x_0$ is the developing point. 
- At the developing point, the Taylor polynomial is exactly the function: $f(x)=T_n(x,x_0)$
-Otherwise, it approximates the function: $f(x)\simeq T_n(x,x_0)$.

For $x_0=0$, the Taylor polynomial reduces to the **McLaurin series**:
$$
T_n(x,0) 
= \sum\limits_{n=0}^{\infty} \frac{f^{(n)}(0)}{n!} x^n
$$
