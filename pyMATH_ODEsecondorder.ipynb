# `pyMATH`

pyMATH, a program package for playing with math in python.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
```

----
## Ordinary differential equations

----
### Second order

We continue with an **ordinary** differential equation of **second** order:
$$
y'' = -y
$$
with 
- $y(t)$ the function of the independent variable $t$, and 
- $y''$ the short-hand notation of $y''=\frac{\partial^2 y}{\partial t^2}$.

This is a second-order ODE, because
- it depends on **one** variable,
- it has **two** derivatives,
- it is **one** equation.

To solve this second-order DGL numerically, we transform it into two 
first-order ODEs, using $\frac{\partial y}{\partial t}=z$.
$$\begin{array}{rcl}
y' &=& z \\
z' &=& -y
\end{array}
$$

We now have a **system of ODEs** of first-order, which can be solved, 
subject to two **initial conditions**:
$$\begin{array}{rcl}
y(t=0) &=& y_0 \\
z(t=0) &=& z_0
\end{array}
$$


A **solution** of this ODE is the **sinus or the consinus function**:
$$
y(t) = \left\{
\begin{array}{c}
\sin(t) \\ \cos(t)
\end{array}
\right.
$$
We test this with the `solve_ivp` solver from the `scipy` package:

```python
# set variable range
t = np.linspace(0,6,51)

# define the right-hand side
def rhs(t,y):
    dydt = y[1]
    dzdt = -y[0]
    return [dydt,dzdt]

# define the initial condition
y0 = [0,1]


# solve the ODE for the given range
dydt = scipy.integrate.solve_ivp(rhs,[t.min(),t.max()],y0,t_eval=t,args=())
#print(dydt.t,dydt.y)
```

```python
plt.figure(figsize=(12,6))
plt.plot(t,np.sin(t),linewidth=10,color='gray',label='$f(t)=\sin(t)$')
plt.plot(dydt.t,dydt.y[0],linewidth=2,color='orange',label='ode solution')
plt.grid()
plt.legend()
```
