## Complex exponential function

We define the complex exponential function 
$$
f(t)=e^{it}
$$

Using the Taylor series expansion (see McLaurin series at end), we recast the series into real and imaginary parts:
$$
\begin{array}{rcl}
e^{it} 
= \sum\limits_{n=0}^{\infty} \frac{(it)^n}{n!} 
&=& 1 + it + \frac{(it)^2}{2!} + \frac{(it)^3}{3!} + \frac{(it)^4}{4!} + \frac{(it)^5}{5!} + \dots \\
&=& 1 + it - \frac{t^2}{2!} - i \frac{t^3}{3!} + \frac{t^4}{4!} - i \frac{t^5}{5!} \pm \dots \\
&=& \left( 1 - \frac{t^2}{2!} + \frac{t^4}{4!} \pm \dots \right) + i \left( t - \frac{t^3}{3!} - \frac{t^5}{5!} \pm \dots \right) \\
&=& \cos(t) + i \sin(t)
\end{array}
$$

$$\fbox{$
e^{it} 
= \cos(t) + i \sin(t)
$}$$

## Complex cosine and sine functions

$$
\begin{array}{rcl}
\frac{1}{2} \left( e^{it} + e^{-it} \right)
&=& \frac{1}{2} \left[ \cos(t) + i \sin(t) + \cos(-t) + i \sin(-t) \right] \\
&=& \frac{1}{2} \left[ \cos(t) + i \sin(t) + \cos(t) - i \sin(t) \right] \\
&=& \cos(t)
\end{array}
$$

$$
\begin{array}{rcl}
\frac{1}{2} \left( e^{it} - e^{-it} \right)
&=& \frac{1}{2} \left[ \cos(t) + i \sin(t) - \cos(-t) - i \sin(-t) \right] \\
&=& \frac{1}{2} \left[ \cos(t) + i \sin(t) - \cos(t) + i \sin(t) \right] \\
&=& i \sin(t)
\end{array}
$$

Rearranging (use $\frac{1}{i}=-i$):
$$\fbox{$
\begin{array}{rcl}
\cos(t) &=& \frac{1}{2} \left( e^{it} + e^{-it} \right) \\
\sin(t) &=& \frac{1}{2i} \left( e^{it} - e^{-it} \right)
\end{array}
$}$$
